```python
myquery = {"MatchHistory": {"$regex": "^(http|https)://matchhistory"}}
mydoc = mycol.find(myquery)
```

```python
nb_error = 0

for doc in mydoc:
    try:
        server = doc["MatchHistory"].split("/")[5]
        gameID = doc["MatchHistory"].split("/")[6].split("?gameHash=")[0]
        hash = doc["MatchHistory"].split("/")[6].split("?gameHash=")[1].split("&")[0]
        url = base_match_history_stats_url.format(server, gameID, hash)
        timeline_url = base_match_history_stats_timeline_url.format(server, gameID, hash)
        # print(url)
        # print(timeline_url)
        game_data = requests.get(url, cookies={c.split("=")[0]:c.split("=")[1] for c in cookie.split(";")}).json()
        game_data_timeline = requests.get(timeline_url, cookies={c.split("=")[0]:c.split("=")[1] for c in cookie.split(";")}).json()
        x1 = mycol1.insert_one(game_data)
        x2 = mycol2.insert_one(game_data_timeline)
    except:
        print("Error parsing url : "+url+"\n"+timeline_url)
        nb_error += 1

print(nb_error)
```
